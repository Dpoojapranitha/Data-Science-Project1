```python
#Matplotlib-legends
import matplotlib.pyplot as plt
year=[1,2,3,4]
Veg_rate=[2,4,5,7]
Milk_rate=[1,8,5,8]
Fruit_rate=[3,2,5,9]
line1, = plt.plot(year,Veg_rate)
line2, = plt.plot(year,Milk_rate)
line3, = plt.plot(year,Fruit_rate)
plt.legend([line1, line2, line3],['Vegetable','Milk','Fruits'])
plt.xlabel('Years')
plt.ylabel('Rate')
plt.show()
print("Successfully Placed a legend on the axis..")
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['Lion','Horse','Kangaroo']
values1=[15,24,30]
values2=[20,18,25]
bar_width=0.35
bar_position1=np.arange(len(categories))
bar_position2=bar_position1+bar_width
plt.bar(bar_position1,values1,width=bar_width,label='2021',color='skyblue')
plt.bar(bar_position2,values2,width=bar_width,label='2022',color='yellow')
plt.xlabel('ctaegories')
plt.ylabel('values')
plt.title('Grouped bar graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=['Transport','Income Tax','Housing']
values1=[15,24,30]
values2=[20,18,25]
plt.bar(categories,values1,label='First Half',color='skyblue')
plt.bar(categories,values2,bottom=values1,label='Second Half')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
exam_scores=np.random.randint(0,100,150)
plt.hist(exam_scores,bins=20,cumulative=True,color='Blue')
plt.xlabel("Exam Scores")
plt.ylabel("Cumulative Number of Students")
plt.title("Cumulative Histogram of Exam Students")
plt.show()
```

```python
#Stacked Histogram
import matplotlib.pyplot as plt
import numpy as np
mark1=np.array([2,4,5,7,9,10,11,13,14,15])
mark2=np.array([6,7,8,10,11,12,13,14,15,16])
plt.hist([mark1,mark2],bins=10,stacked=True,color=['Blue','Salmon'],edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('Stacked Histogram with Multiple Datasets')
plt.legend(['Mark Dataset 1','Mark Dataset2'])
plt.show()
```
